```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline     
sns.set(color_codes=True)
```

```python
#loeading to DF
df = pd.read_csv("data.csv")
# To display the top 5 rows 
df.head(5)  
```

```python
#Remove irrelevant columns using "df.drop"
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

```python
# Rename columns
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
df.head(5)
```

```python
#Check for duplicates
df.shape
```

```python
#Create df of dupicates to count how many
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
#remove duplicates
df = df.drop_duplicates()
df.head(5)
```

```python
#count after removing duplicates
df.count()
```

```python
#Treat missin/null values - can be seen above in the df.count()
# removed in this case. Could be replaced with average or other

print(df.isnull().sum())
```

```python
# Dropping the missing values - all equal now
df = df.dropna()    
df.count()
```

```python
# After dropping the values - to check
print(df.isnull().sum())   
```

Detecting Outliers - can be very high or very low. Good idea to detect and remove the outliers because outliers are one of the primary reasons for resulting in a less accurate model

Often outliers can be seen with visualizations using a box plot (1.5xinterquartile range)



```python
sns.boxplot(x=df['Price'])
```

```python
sns.boxplot(x=df['HP'])
```

```python
sns.boxplot(x=df['Cylinders'])
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# Remove outliers >1.5x IQR
# ~ is NOT operator. | is OR operator. & is AND operator
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
#Histogram 
df.Make.value_counts().plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

```python
#Heatmaps
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
#Scatterplot 
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['HP'], df['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```
